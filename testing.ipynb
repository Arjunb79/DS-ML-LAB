```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
s=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/sonar data.csv')
```

```python
scaler=StandardScaler()
scaler.fit(s.drop('R',axis=1))
s.columns[:-1]
```

```python
scaled_features=scaler.transform(s.drop('R',axis=1))
scaled_features
```

```python
new_data=pd.DataFrame(scaled_features,columns=s.columns[:-1])
new_data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix,classification_report
```

```python
x=new_data
y=s['R']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(x_train,y_train)
pred=knn.predict(x_test)
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

```python
error_rate=[]
for i in range(1,40):
  knn=KNeighborsClassifier(n_neighbors=i)
  knn.fit(x_train,y_train)
  pr=knn.predict(x_test)
  error_rate.append(np.mean(pr!=y_test))
error_rate  

```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue')
```

```python
knn1=KNeighborsClassifier(n_neighbors=4)
knn1.fit(x_train,y_train)
pred=knn1.predict(x_test)
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
data=load_iris()
data.data.shape
data.target.shape
```

```python
x=data.data
y=data.target
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=50)

de=DecisionTreeClassifier(criterion='gini')
de.fit(x_train,y_train)
pred=de.predict(x_test)
print('trained accuracy',accuracy_score(y_train,de.predict(x_train)))
print('test accuracy',accuracy_score(y_test,pred))
```

```python
from sklearn.tree import export_graphviz
from six import StringIO
from IPython.display import Image
import pydotplus

dot_data=StringIO()
export_graphviz(de,out_file=dot_data,filled=True,rounded=True,special_characters=True,
                feature_names=data.feature_names,class_names=data.target_names)
graph=pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```
